```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
import numpy as np
import os
apath = '/Users/richardanderson/Downloads/planetesimal_project/shusolutionoutput/alpha'
alpha = {}
for filename in os.listdir(apath):
    if filename.endswith('.npy'):
        file_path = os.path.join(apath, filename)
        array = np.load(file_path)
        key = os.path.splitext(filename)[0]
        alpha[key] = array
```

```python
x = np.load('/Users/richardanderson/Downloads/planetesimal_project/shusolutionoutput/x/x2.1.npy')
```

```python
%matplotlib widget

plt.clf()
plt.plot(x, alpha['alpha2.1'], label = '$A = 2.1$')
plt.plot(x, alpha['alpha2.000000000001'], label = '$A = 2.000000000001$')
plt.xlabel('$x$')
plt.ylabel('$\\alpha$')
plt.legend()
plt.yscale("log")
plt.show()
```
