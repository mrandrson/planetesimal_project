```python
%matplotlib widget

import os
import matplotlib.pyplot as plt
import pandas as pd

plt.clf()

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize = (12, 8))

for f in os.listdir('.'):
    if f.endswith('.csv'):
        dat = pd.read_csv(f)
        for p in dat['particle'].unique():
            pdat = dat[dat['particle'] == p]

            t = pdat['time']
            e = pdat['energy']
            r = pdat['r']
            phi = pdat['potential']
            v = pdat['v']
            KE = 0.5*v**2
            ax1.plot(t, r)
            ax1.set_yscale('symlog')
            ax1.set_ylabel(r"$r\ [m]$")
            ax2.plot(t, phi)
            ax2.set_ylabel(r"$\varphi\ \left[\frac{m^2}{s^2}\right]$")
            ax2.set_yscale('symlog')
            ax3.plot(t, KE)
            ax3.set_ylabel(r'$E_k\ \left[\frac{m^2}{s^2}\right]$')
            ax3.set_xlabel(r'$t\ [yr]$')
            ax3.set_yscale('log')


plt.show()
```
